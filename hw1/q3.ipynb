```python
def chart(default, date1, date2):
    if(default == 1):
        df.Close.plot(figsize=(16,8), title="Apple")
        plt.show()
    else:
        mask = (df['Date'] > date1) & (df['Date'] <= date2)
        df2 = df.loc[mask]
        df2.Close.plot(figsize=(16,8), title="Apple")
        plt.show()
```

```python
chart(0, "2014-6-1", "2014-6-10")
```

```python
rolling_mean = df.Close.rolling(window=3).mean()
plt.plot(df.Date, df.Close, label='APPLE')
plt.plot(df.Date, rolling_mean, label='AMD 3 Day SMA', color='orange', alpha=0.5)
plt.legend(loc='upper left')
plt.show()
```
